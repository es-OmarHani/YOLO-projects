# Make data yaml file  

```python
yaml_content = """
# Data
path: /content/drive/MyDrive/data
train: /content/drive/MyDrive/data/images/train  # train images (relative to 'path')
val: /content/drive/MyDrive/data/images/val  # val images (relative to 'path')

# Keypoints
kpt_shape: [39, 3]  # [number of keypoints, number of dim]
flip_idx: [0, 1, 2, 4, 3, 10, 11, 12, 13, 14, 5, 6, 7, 8, 9, 15, 16, 17, 18, 19, 20, 21, 22, 23, 27, 28, 29, 24, 25,
           26, 33, 34, 35, 30, 31, 32, 36, 38, 37]

# Classes
names:
  0: quadruped
"""

# Write the content to a YAML file named 'data.yaml'
with open('data.yaml', 'w') as yaml_file:
    yaml_file.write(yaml_content)
```

# Train yolo_v8 on our Data"

```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolov8n-pose.pt")  # load a pretrained model (recommended for training)

# Use the model
model.train(data="data.yaml", epochs=100, imgsz=640)  # train the model
# metrics = model.val()  # evaluate model performance on the validation set
```
